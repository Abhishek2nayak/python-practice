1 > Write a python script to print country and their capital names from
the dictionary.

```python
countries = {
    'US': {'capital': 'Washington D.C.', 'car_per_cap': 809, 'population': 46.77},
    'Australia': {'capital': 'Vienna', 'car_per_cap': 731, 'population': 66.03},
    'Japan': {'capital': 'Tokyo', 'car_per_cap': 588, 'population': 80.2},
    'Russia': {'capital': 'Moscow', 'car_per_cap': 18, 'population': 5.084},
    'India': {'capital': 'New Delhi', 'car_per_cap': 18, 'population': 5.084},
    'Morocco': {'capital': 'Rabat', 'car_per_cap': 18, 'population': 55.4}
}


for k , v in countries.items():
    print(k," ==> ",v["capital"])

```

2 > Find top 3 countries having highest population and print the same.

```python
sorted_c = sorted(countries.items(), key = lambda x : x[1]["population"] , reverse=True)
sorted_dict = dict(sorted_c)
c = 0
for k, v in sorted_dict.items():
    if c < 3 :
        print(k," ==>  ",v['population'])
    c = c+ 1    

    

```

3 > Write a python script to sort a dictionary by car_per_cap print the
Country, Capital and Car_Per_Ca

```python
sorted_c = dict(sorted(countries.items(), key = lambda x : x[1]['car_per_cap'],reverse=True ))


for k , v in sorted_c.items():
    print(k," => ",v['capital'],' - ',v['car_per_cap'])
```

4 > Plot bar graph to display country-wise population ratio

```python
import matplotlib.pyplot as pt

country = []
popu = []
for k, v in countries.items():
    country.append(k)
    popu.append(v['population'])

pt.bar(country,popu)
pt.show()

pt.pie(popu,labels=country,autopct = "%1.1f%%")
pt.legend(title = 'Countries')
pt.show()
```
